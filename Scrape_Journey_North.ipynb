```python
import pandas as pd
```

```python
# First egg sightings
url = "https://journeynorth.org/sightings/querylist.html?season=spring&map=monarch-egg-first&year=2021&submit=View+Data"
```

```python
# Scrape egg sightings and read to dataframe
eggs_spring_2021 = pd.read_html(url)
eggs_spring_2021 = eggs_spring_2021[0]
eggs_spring_2021
```

```python
# Drop Image and Number columns
eggs_spring_2021.drop(columns=['Number', 'Image', 'Unnamed: 0'], inplace=True)
eggs_spring_2021
```

```python
# Add identifying column
eggs_spring_2021.insert(1, "Type", "egg")
eggs_spring_2021
```
